```python
import pandas as pd
```

```python
datasetPath = '/content/drive/MyDrive/retail_transactions.csv'
df = pd.read_csv(datasetPath)
```

```python
df['TransactionDate']=pd.to_datetime(df['TransactionDate'])
```

```python
df['Quantity']=pd.to_numeric(df['Quantity'])
df['Price']=pd.to_numeric(df['Price'])
df['Discount']=pd.to_numeric(df['Discount'])

df['Discount']=df['Discount'].fillna(0)
```

```python
df
```

```python
df['NetPrice']=df['Price']* (1 - df['Discount'] / 100)
```

```python
df['NetPrice']
```

```python
df['Totalsale'] = df['Quantity'] * df['NetPrice']
```

```python
df['Totalsale']
```

```python
filtered_df = df[df['Totalsale'] >= 20]
```

```python
filtered_df
```

```python
df1 = filtered_df
```

```python
df1
```

```python
df['PaymentType']= df['PaymentType'].astype('category')
```

```python
df['PaymentType']
```

```python
grouped_df = df1.groupby('TransactionDate').sum().reset_index()
```

```python
grouped_df
```

```python
plt.bar(grouped_df.index,grouped_df['Totalsale'],color = 'red')
plt.xlabel('product category')
plt.ylabel('total sales')
plt.title('total sales per product category')
plt.show()
```
