```python
def AnalyzeTowns():
  TownData=df['home_team'].value_counts()
  for _ in TownData.index:
    print(f'Total Number of matches played by{_}is :{TownData[_]}')
```

```python
townName =[]
matchesplayed=[]
TownData = df['home_team'].value_counts()
for _ in TownData.index:
  if TownData[_] > 30 :
    townName.append(_)
    matchesplayed.append(TownData[_])
```

```python
from matplotlib import pyplot as plt
```

```python
plt.bar(matchesplayed ,townName)
plt.xlabel('Town Name')
plt.ylabel('Matches Played')
plt.title('Matches Played by Towns')
plt.show()
```

```python
def MatchesplayedByTown():
  townName = []
  matchesplayed=[]
  TownData = df['home_team'].value_counts()
  for _ in TownData.index:
    if TownData[_] > 30 :
      townName.append(_)
      matchesplayed.append(TownData[_])
  plt.plot(matchesplayed ,townName)
  plt.xlabel('Town Name')
  plt.ylabel('Matches Played')
  plt.title('Matches Played by Towns')
  plt.show()
```

```python
MatchesplayedByTown()
```
